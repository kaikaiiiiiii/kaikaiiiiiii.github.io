```python
import matplotlib.pyplot as plt
import numpy as np


def n_devide_circle(n):
    """
    """

    # 圆的参数
    radius = 0.5
    center = (0, 0)

    # 生成圆周上的点
    angles = np.linspace( 2 * np.pi+np.pi*(1/2-2/n), np.pi*(1/2-2/n) , n + 1)  # 8个点，包含从圆的上顶端开始
    x = center[0] + radius * np.cos(angles)
    y = center[1] + radius * np.sin(angles)

    # 创建图形和轴
    fig, ax = plt.subplots()

    # 绘制圆
    circle = plt.Circle(center, radius, color='blue', fill=False, linestyle='--')
    ax.add_artist(circle)

    # 绘制圆周上的点和标签
    # if n > 7 then dont write lable
    for i in range(n):
        ax.plot(x[i], y[i], 'ro')  # 绘制红色点
        if n <= 7:
            label = f"{(i + 1) / n:.{n}f}"[:n+1]  # 格式化标签
            ax.text(x[i]-0.15, y[i]-0.1, label, fontsize=12, ha='left')

    # 设置轴的比例和范围
    ax.set_aspect('equal')
    ax.set_xlim(center[0] - radius - 0.1, center[0] + radius + 0.1)
    ax.set_ylim(center[1] - radius - 0.1, center[1] + radius + 0.1)

    # 移除坐标轴
    ax.axis('off')

    # 添加网格和标题
    plt.grid(True)
    plt.title("")

    # 显示图形
    plt.show()


n_devide_circle(7)
```
