```python
import matplotlib.pyplot as plt
```

```python
def eularShieve(n):
    prime = []
    isPrime = [True for i in range(n+1)]
    isPrime[0] = isPrime[1] = False
    for i in range(2, n+1):
        if isPrime[i]:
            prime.append(i)
            for j in range(i*i, n+1, i):
                isPrime[j] = False
    return prime
```

```python
primes = eularShieve(10**8)
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

def count_ranges(array, k):
    counts = []
    start = 1
    end = k
    i = 0
    while i < len(array):
        count = 0
        while i < len(array) and array[i] <= end:
            count += 1
            i += 1
        counts.append(count)
        start = end + 1
        end = (start - 1) + k
    return counts

k=100000

# 设置图片宽度为三倍
plt.figure(figsize=(24, 5))

# 计算 y 值
y_values = count_ranges(primes, k)

# 计算 x 值
x_values = [i * k for i in range(1, len(y_values) + 1)]


# 绘图
plt.plot(x_values, y_values, marker='.', markersize=2, color='green', label='Count',linestyle='None')
plt.text(x_values[0]+300000, y_values[0], f'{y_values[0]}', fontsize=9, verticalalignment='center', horizontalalignment='left', color='green')
plt.plot(x_values[0], y_values[0], marker='o', markersize=3, color='green', label='Count',linestyle='None')
plt.xlabel(f'Interval {k}')
plt.xlim(-1, 1e8)
plt.xticks(range(0, 10**8+1000, 10**7))
plt.ylabel('Count')
plt.title(f'Primes count of every {k} numbers')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.grid(True)
plt.show()
```
