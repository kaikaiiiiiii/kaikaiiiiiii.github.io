```python
import matplotlib.pyplot as plt
```

```python
def factorial(n):
   
    f = 0
    for i in range(1,n+1):
        c = 1
        c = c * n**(n-i) * i**2
        for j in range(1,i):
            c = c * (n-j)
        s = 1
        s=s * c
        f = f + s
    base=n**n

    return base/f
```

```python
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

def remove_common_factors(A, B):
    common_factor = gcd(A, B)
    a = A // common_factor
    b = B // common_factor
    return a, b
#remove_common_factors(base,f)
```

```python
# 初始化空列表用于存储(x, y)点
points = []

# 循环输入x的值，计算对应的y值，并将(x, y)点添加到列表中
for x in range(1, 104):
    y = factorial(x)
    points.append((x, y))

# 提取x和y坐标
x_coords = [point[0] for point in points]
y_coords = [point[1] for point in points]

# 创建一个新的图形
plt.figure(figsize=(20, 5))

# 绘制(x, y)点，调整点的大小
plt.scatter(x_coords, y_coords, s=10, color='#993285')  # s参数控制点的大小

# 添加标题和标签
# plt.title('Q(n)')
# plt.xlabel('n')
# plt.ylabel('Q(n)')

# 设置x轴范围从0开始
plt.xlim(0,103)
plt.ylim(0,1.1)

# 添加坐标轴背景
plt.grid(which='both', axis='both', color='gray', linestyle='--', linewidth=0.5, alpha=0.5)
plt.xticks(range(0, 101, 5))
plt.yticks([i/10 for i in range(0, 11)])

# 保存图片
plt.savefig('plot.png')

# 显示图形
plt.show()
```
