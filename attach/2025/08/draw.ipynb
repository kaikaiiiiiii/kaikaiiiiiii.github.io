```python
def plot_factor_product(x_start=1, x_end=None, relative=False, printValue=False):
    n = 10**6

    # 如果未指定x_end，则使用完整范围
    if x_end is None:
        x_end = n
    else:
        x_end = min(x_end, n)

    # 初始化数据结构
    spf = np.zeros(n + 1, dtype=np.uint32)  # 最小质因数数组
    primes = []  # 素数列表
    prime_index = {}  # 素数索引映射
    product = np.zeros(n + 1, dtype=np.uint32)  # 因子序数乘积数组

    # 构建SPF和素数序号
    for i in range(2, n + 1):
        if spf[i] == 0:
            spf[i] = i
            primes.append(i)
            prime_index[i] = len(primes)  # 记录素数序号

        j = 0
        while j < len(primes):
            p = primes[j]
            ip = i * p
            if p > spf[i] or ip > n:
                break
            spf[ip] = p
            if i % p == 0:
                break
            j += 1

    # 计算因子序数乘积
    product[1] = 1
    for i in range(2, n + 1):
        p = spf[i]  # 最小质因数
        prev = i // p
        product[i] = product[prev] * prime_index[p]

    # 截取指定范围的数据
    x = np.arange(x_start, x_end + 1)
    y = product[x_start : x_end + 1]

    # 应用相对值模式
    if relative:
        y = y / x

    # 创建绘图
    plt.figure(figsize=(14, 8), dpi=100)

    # 绘制点图
    plt.scatter(x, y, s=1, color="blue", alpha=0.7)

    # 添加数值标注
    if printValue:
        # 获取当前y轴范围来计算合适的偏移量
        y_min, y_max = plt.ylim()
        # 计算偏移量 - 使用y轴范围的1%作为偏移量
        offset = (y_max - y_min) * 0.01

        # 为了避免重叠，只标注较大的点
        data_max = np.max(y)
        for i, val in zip(x, y):
            # if val > 0.1 * data_max:  # 只标注大于10%最大值的点
                plt.text(
                    i,  # x坐标不变
                    val + offset,  # 在原始y值上增加偏移量
                    f"{i}",
                    rotation=90,
                    fontsize=8,
                    ha="center",
                    va="bottom",  # 文字底部对齐点位置
                    alpha=0.75,
                )

    # 设置坐标轴范围
    plt.xlim(x_start - 0.5, x_end + 0.5)

    if relative:
        plt.ylim(0, np.max(y) * 1.1)
    else:
        plt.ylim(0, np.max(y) * 1.05)

    # 设置刻度
    max_x_ticks = min(12, x_end - x_start + 1)
    max_y_ticks = 8

    # X轴刻度（等间距）
    x_ticks = np.linspace(x_start, x_end, max_x_ticks, dtype=int)
    plt.xticks(x_ticks)

    # Y轴刻度（等间距）
    y_min, y_max = plt.ylim()
    y_ticks = np.linspace(y_min, y_max, max_y_ticks)
    plt.yticks(y_ticks)

    # 标签和标题
    # plt.xlabel("i", fontsize=12)

    # if relative:
        # plt.ylabel("product[i] / i", fontsize=12)
        # plt.title(
        #     f"Factor Index Product Relative Distribution ({x_start} to {x_end})",
        #     fontsize=14,
        # )
    # else:
        # plt.ylabel("product[i]", fontsize=12)
        # plt.title(
        #     f"Factor Index Product Distribution ({x_start} to {x_end})", fontsize=14
        # )

    # 移除网格
    plt.grid(False)
    plt.tight_layout()

    # 显示图形
    plt.show()


# 示例调用
# plot_factor_product()  # 完整范围，绝对数值
# plot_factor_product(relative=True)  # 完整范围，相对数值
plot_factor_product(29000, 30000, relative=False, printValue=True)  # 局部范围，带数值标注
```
